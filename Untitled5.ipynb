```python
# Importar las librerías necesarias
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor

# Datos del ejemplo (Tamaño de la casa y Precio de venta)
house_size = [
    52, 65, 60, 85, 70, 110, 120, 95, 80, 55, 90, 130, 75, 140,
    50, 100, 160, 125, 145, 140, 115, 130, 180, 165, 135, 115,
    100, 190, 75, 105, 80, 125, 150, 160, 58, 62, 92, 67,
    57, 69, 72, 128, 78, 99, 56, 88, 116, 139, 53, 76,
    150, 66, 63, 161, 94, 97, 114, 135, 121, 54, 83, 149,
    173, 71, 84, 123, 76, 110, 140, 135, 102, 99, 170, 81,
    64, 125, 78, 118, 161, 148, 119, 112, 69, 72, 53, 107,
    161, 86, 150, 162, 150, 70, 63, 65, 73, 168, 162, 85,
    77, 150, 164, 82, 95, 130, 108, 155, 140, 155, 149, 70
]

price = [
    155000, 180500, 175000, 225000, 210000, 290000, 315000, 245000,
    235000, 160000, 255000, 340000, 320000, 365000, 150000, 275000,
    400000, 330000, 370000, 355000, 300000, 335000, 450000, 410000,
    350000, 290000, 275000, 290000, 220000, 285000, 230000, 310000,
    180000, 200000, 158000, 170000, 260000, 180000, 162000, 175000,
    200000, 330000, 210000, 270000, 157000, 250000, 295000, 360000,
    156000, 230000, 385000, 180000, 172000, 250000, 258000, 265000,
    292000, 345000, 315000, 155000, 220000, 375000, 425000, 205000,
    227000, 322000, 218000, 290000, 360000, 340000, 280000, 272000,
    420000, 225000, 178000, 312000, 211000, 300000, 405000, 373000,
    295000, 280000, 185000, 202000, 157000, 285000, 410000, 230000,
    380000, 390000, 200000, 175000, 179000, 210000, 430000, 410000
]

# Asegurarse de que ambas listas tienen el mismo tamaño
if len(house_size) < 100:
    additional_house_sizes = [80, 90, 100, 110, 120] * 20
    house_size.extend(additional_house_sizes[:100 - len(house_size)])

if len(price) < 100:
    additional_prices = [250000, 270000, 290000, 310000, 330000] * 20
    price.extend(additional_prices[:100 - len(price)])

house_size = house_size[:100]
price = price[:100]

assert len(house_size) == len(price), "Las listas deben tener la misma longitud."

# Ordenar los datos de menor a mayor
sorted_data = sorted(zip(house_size, price))
house_size, price = zip(*sorted_data)

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(np.array(house_size).reshape(-1, 1), price, test_size=0.2, random_state=42)

# Crear el modelo de Gradient Boosting
model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
model.fit(X_train, y_train)

# Realizar predicciones
y_pred = model.predict(X_test)

# Configurar la figura para visualizar los datos y la predicción
fig, ax = plt.subplots(figsize=(10, 6))

# Graficar los datos originales
ax.scatter(house_size, price, color='blue', label='Datos reales')

# Graficar las predicciones del modelo
ax.scatter(X_test, y_pred, color='red', label='Predicciones del modelo', marker='x')

# Trazar la línea de regresión del modelo
x_range = np.linspace(50, 200, 100).reshape(-1, 1)
y_range = model.predict(x_range)
ax.plot(x_range, y_range, color='green', label='Modelo de Gradient Boosting')

# Función para dibujar un árbol de decisión en la gráfica
def draw_tree(ax, x, y, width, height, level):
    # Dibujar un rectángulo que representa el árbol
    ax.add_patch(plt.Rectangle((x - width / 2, y), width, height, edgecolor='black', facecolor='orange'))
    # Agregar un texto para el árbol
    ax.text(x, y + height + 0.02, f'Tree {level}', ha='center')

# Dibujar árboles de decisión
num_trees = 5  # Número de árboles a dibujar
for i in range(num_trees):
    draw_tree(ax, x=2 + i * 2, y=y_range[i * 20 + 10], width=1, height=0.1, level=i + 1)

# Añadir etiquetas y título
ax.set_xlabel('Tamaño de la Casa (m²)')
ax.set_ylabel('Precio (USD)')
ax.set_title('Predicción de Precio de Casa usando Gradient Boosting con Árboles de Decisión')
ax.legend()

# Mostrar la gráfica
plt.show()
```
