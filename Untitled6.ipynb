```python
import numpy as np
import matplotlib.pyplot as plt

# Generar datos sintéticos
np.random.seed(0)
data_class1 = np.random.randn(50, 2) + np.array([5, 5])
data_class2 = np.random.randn(50, 2) + np.array([-5, -5])
X = np.vstack([data_class1, data_class2])

# Función para calcular la distancia euclidiana
def euclidean_distance(a, b):
    return np.sqrt(np.sum((a - b) ** 2))

# Implementación del algoritmo K-Means
def k_means(X, k, max_iters=100):
    # Inicializar los centroides aleatoriamente
    indices = np.random.choice(X.shape[0], k, replace=False)
    centroids = X[indices]

    for _ in range(max_iters):
        # Asignar cada punto al centroide más cercano
        clusters = [[] for _ in range(k)]
        for x in X:
            distances = [euclidean_distance(x, centroid) for centroid in centroids]
            cluster_index = np.argmin(distances)
            clusters[cluster_index].append(x)

        # Actualizar los centroides
        new_centroids = np.array([np.mean(cluster, axis=0) if len(cluster) > 0 else centroids[i]
                                  for i, cluster in enumerate(clusters)])

        # Si los centroides no cambian, salir del ciclo
        if np.all(centroids == new_centroids):
            break

        centroids = new_centroids

    return centroids, clusters

# Ejecutar K-Means con k=2 (dos clusters)
k = 2
centroids, clusters = k_means(X, k)

# Visualizar los resultados
plt.figure(figsize=(10, 6))
colors = ['blue', 'red']
for i, cluster in enumerate(clusters):
    cluster = np.array(cluster)
    plt.scatter(cluster[:, 0], cluster[:, 1], color=colors[i], label=f'Cluster {i+1}')
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', s=100, label='Centroids')
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("K-Means Clustering")
plt.legend()
plt.show()
```
